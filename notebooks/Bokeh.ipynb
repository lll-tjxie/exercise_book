```python
from bokeh.charts import Scatter, output_notebook, show
import pandas as pd
```

```python
df = pd.DataFrame(columns=['X','Y'])
df['X'] = [1,2,3,4,5]
df['Y'] = [6,4,2,9,5]
df
```

```python
from bokeh.charts import Scatter, output_notebook, show
p = Scatter(df,x='X',y='Y', \
           title="Temperature Observations", \
           xlabel="Day of observation",
           ylabel="Temperature",
           tools = 'pan,resize')
p.toolbar.logo = None

output_notebook()
show(p)
```

```python
from bokeh.charts import output_file

p = Scatter(df,x='X',y='Y', \
           title="Temperature Observations", \
           xlabel="Day of observation",
           ylabel="Temperature",
           tools = 'pan,resize')
p.toolbar.logo = None

output_file('scatter_charts.html')
show(p)
```

```python
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=400, plot_height=400, 
           title='Temperature Observation')
#p.circle([1,2,3,4,5], [9,2,6,2,5], size=12, color='red', alpha=0.5)
p.circle([1,2,3,4,5], [9,2,6,2,5], 
         #size=[6,9,12,15,18],
         size=[i*2 for i in [6,9,12,15,18]],
         color='red', alpha=0.5)
p.triangle([1,2,3,4,5], [1,9,4,8,2], 
         #size=[6,9,12,15,18],
         size=[i*2 for i in [6,9,12,15,18]],
         color='green', alpha=0.5)
output_file("scatter_plotting.html")
show(p)
```

```python
p = figure(plot_width=400, plot_height=400)

p.title.text = 'Temperature Observations'
p.title.text_font = "times"
p.title.text_font_style = "italic"

p.circle([1,2,3,4,5], [9,2,6,2,5], size=12, color='red', alpha=0.5)

output_file('scatter_plotting.html')
show(p)
```

```python
p = figure(plot_width=400, plot_height=400)

p.title.text = 'Temperature Observations'
p.title.text_font = "times"
p.title.text_color = 'blue'
p.title.text_font_style = "italic"

p.xaxis.axis_label = 'Day of Observation'
p.xaxis.minor_tick_line_color = None
p.yaxis.axis_label = 'Temp'

p.circle([1,2,3,4,5], [9,2,6,2,5], size=12, color='red', alpha=0.5)
p.line([1,2,3,4,5], [2,5,20,0.5,9],line_width=5, color='blue', alpha=0.5)

output_file('scatter_plotting.html')
show(p)
```

```python
from bokeh.plotting import figure, output_notebook, show, gridplot
from numpy import *

x = arange(0,2*pi,0.1)

g1 = figure(width=400, height=400, title=None)
g1.line(x, sin(x), legend = 'sine of x')

g2 = figure(width=400, height=400, title=None)
g2.line(x, cos(x), legend = 'cosine of x')

gdp = gridplot([[g1,g2]]) # if ([[g1,g2],[g3,g4]]), g1g2 will be on first layer g3g4 on second...

output_notebook()
show(gdp)
```
