```python
import pandas as pd
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data'
col_names = ['id','ri','na','mg','al','si','k','ca','ba','fe','glass_type']
glass = pd.read_csv(url, names = col_names, index_col=0)
glass
```

```python
dic = {1:0,2:0,3:0,4:0,5:1,6:1,7:1}
```

```python
glass['assorted'] = glass.glass_type.map(dic)
glass
```

```python
import sklearn.linear_model
```

```python
lr = sklearn.linear_model.LogisticRegression(C=1e9)
X = glass[['al']]
y = glass.assorted
lr.fit(X,y)
predictions = lr.predict(X)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.scatter(X, predictions,color = 'red')
```

```python
proba = lr.predict_proba(X)
proba
```

```python
plt.scatter(X, y)
plt.scatter(X, proba[:,0])
plt.scatter(X, proba[:,1], color = 'red')
```

```python
# Al = 2
logodds = lr.intercept_ + lr.coef_*2
```

```python
import numpy as np
odds = np.exp(logodds)
```

```python
probability = odds/(1+odds)
probability
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

plt.style.use('seaborn-white')
%matplotlib inline
```

```python
# generic curve plotting function
def auc_plotting_function(rate1, rate2, rate1_name, rate2_name, curve_name):
    AUC = auc(rate1, rate2)
    # Plot of a ROC curve for class 1 (has_cancer)
    plt.figure(figsize=[11,9])
    plt.plot(rate1, rate2, label=curve_name + ' (area = %0.2f)' % AUC, linewidth=4)
    plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel(rate1_name, fontsize=18)
    plt.ylabel(rate2_name, fontsize=18)
    plt.title(curve_name + ' for house price > 200,000', fontsize=18)
    plt.legend(loc="lower right")
    plt.show()

# plot receiving operator characteristic curve
def plot_roc(y_true, y_score):
    fpr, tpr, _ = roc_curve(y_true, y_score)
    auc_plotting_function(fpr, tpr, 'False Positive Rate', 'True Positive Rate', 'ROC')
```

```python
sac = pd.read_csv('Sacramentorealestatetransactions.csv')
```

```python
#sac['indicator'] = np.where(sac.price>200000,1,0)
#sac.head()
```

```python
sac['indicator'] = sac['price'].map(lambda x: 1 if x>200000 else 0) 
sac.head()
```

```python
sac = sac[['beds','baths','sq__ft','indicator']]
```

```python
sac
```

```python
X = sac[['beds','baths','sq__ft']]
y = sac.indicator
(X_train,X_test,y_train,y_test) = train_test_split(X,y,test_size = 0.33)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
confusion_dataframe = pd.DataFrame({
        'predictions': y_pred,
        'actuals': y_test
    })
true_positive = ((confusion_dataframe.actuals == confusion_dataframe.predictions) & (confusion_dataframe.predictions==True))
true_negative = ((confusion_dataframe.actuals == confusion_dataframe.predictions) & (confusion_dataframe.predictions==False))
false_positive = ((confusion_dataframe.predictions==True)&(confusion_dataframe.actuals==False))
false_negative = ((confusion_dataframe.predictions==False)&(confusion_dataframe.actuals==True))
print "True positives:", true_positive.sum()
print "True negatives: ", true_negative.sum()
print 'False negatives: ', false_negative.sum()
print 'False positives: ', false_positive.sum()
```

```python
conmat = confusion_matrix(y_test,y_pred)
confusion = pd.DataFrame(conmat, index = ['under_200', 'over_200'], columns = ['pred_under_200', 'pred_over_200'])
print confusion
```

```python
accuracy_score(y_test,y_pred)
```
