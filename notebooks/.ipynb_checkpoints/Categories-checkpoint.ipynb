Adding a column is a category derived from a column that is a string(object)
Category can work with numbers

```python
sacramento['dwelling_type'] = sacramento.type.astype('category')
sacramento
```

convert a category into a several columns of numbers

```python
sacramento_with_dummies = pandas.get_dummies(sacramento,columns = ["dwelling_type"])
sacramento_with_dummies
```

If we attempt to linear regression using all 4 dummy variables, there are an infinite number of options for the zero-intercept point.
sklearn chooses the lowest possible float for this : -1.8485515393235526e+19

```python
import sklearn.linear_model
ols = sklearn.linear_model.LinearRegression()
columns = ['dwelling_type_Condo','dwelling_type_Multi-Family','dwelling_type_Residential','dwelling_type_Unkown']
ols.fit(sacramento_with_dummies[columns],sacramento_with_dummies.price)
zip(columns,ols.coef_)
```

even ignoring one column gives us sensible,comprehensive results.

```python
import sklearn.linear_model
ols = sklearn.linear_model.LinearRegression()
columns = ['dwelling_type_Condo','dwelling_type_Multi-Family','dwelling_type_Residential']
ols.fit(sacramento_with_dummies[columns],sacramento_with_dummies.price)
zip(columns,ols.coef_)
```

sklearn has to choose the zero-intercept to be the mean price of the properties which are all zero in the columns we chose to regress on.

```python
ols.intercept_
```

example of cut and qcut

```python
%matplotlib inline
import seaborn
```

```python
seaborn.distplot(sacramento.sq__ft)
```

```python
sacramento.sq__ft.nlargest(100).min()
```

we can split at 100 per bin from the top, but there would be more than 100 in the bottom bin. we confirm that 200 in the bottom bin is 

```python
sacramento.sq__ft.nsmallest(200).max()
```

```python
pandas.qcut(sacramento.sq__ft,5).value_counts()
```

```python
sacramento['footage_size'] = pandas.qcut(sacramento.sq__ft,5)
```

we could do pretier naming

```python
pandas.qcut(sacramento.sq__ft,5,labels = ['Minimal','Small','Medium','Large','Farm']).value_counts
```

```python
sacramento['footage_description'] = pandas.qcut(sacramento.sq_ft,5,lables = ['Minimal','Small','Medium','Large','Farm'])
```

we can also choose equal-sized footage ranges

```python
pandas.cut(sacramento.sq__ft,10).value_counts()
```

```python
seaborn.distplot(sacramento.sq__ft,10)
```

```python
sacramento['footage_range'] = pandas.cut(sacramento.sq__ft,10)
```
