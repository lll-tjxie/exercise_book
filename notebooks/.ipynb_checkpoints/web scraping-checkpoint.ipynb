```python
wiki = requests.get('https://en.wikipedia.org/wiki/States_and_territories_of_Australia')
```

```python
soup = BeautifulSoup(wiki.text,'lxml')
soup
```

```python
soup.find_all('a')
```

```python
soup('a')
```

```python
soup.find_all('div',class_='toc')
```

```python
def find_the_word_state(x):
    if "state" in x:
        return True
    else:
        return False
soup.find('b',string = find_the_word_state)
```

```python
len(soup.find_all('p'))
```

```python
main_content = soup.find('div', id='mw-content-text')
first_paragraph = main_content.find('p')
first_hyperlink = first_paragraph.find('a')
first_hyperlink
```

```python
first_box = main_content.find('dl')
```

```python
date = first_box.find_all('dd')
date
```
