```python
# create a dataframe with just the estimated price column in it
df = pd.DataFrame({
      'EstimatedPrice' : house_data.EstimatedPrice
    })
df
```

```python
# create a RANSAC regressor
r = sklearn.linear_model.RANSACRegressor()
```

```python
# use the RANSAC regressor to find how much of the Price is related to the EstimatedPrice  
r.fit(df, house_data.Price)
r.estimator_.coef_
```

```python
# use that regression to predict Price based on EstimatedPrice
prediction = r.predict(df[['EstimatedPrice']])
```

```python
# draw a scatter plot of Price / EstimatedPrice and overlay the RANSAC prediction on it
plt.scatter(df.EstimatedPrice, house_data.Price)
plt.scatter(df.EstimatedPrice, prediction,color='red')
```

```python
# RANSAC has an attribute called 'estimator'. What type is it?
type(r.estimator_)
```

```python
# What is the coefficient (gradient) and y-intercept as reported by this estimator (hint: coef_ and intercept_)
r.estimator_.intercept_
```

```python
r.estimator_.coef_
```

```python
# Create a new data frame with two columns: HouseSizeSqft and Bath
#new_df = house_data[['HouseSizeSqft','Bath']].copy()
df2 = pd.DataFrame({
        'HouseSizeSqft':house_data.HouseSizeSqft,
        'Bath':house_data.Bath,
        'Price':house_data.Price
    })
df2
```

```python
df2.dropna(inplace=True)
df2
```

```python
# Regress Price against these values
r2 = sklearn.linear_model.RANSACRegressor()
x = df2[['Bath','HouseSizeSqft']]
y = df2.Price
r2.fit(x,y)
zip(['Bath','HouseSizeSqft'],r2.estimator_.coef_)
```

```python
# What would you expect a house price to be for a house with 3 bathrooms and 3500 sqft?
r2.predict([[3,3500]])
```

```python
prediction = r2.predict(df2[['Bath','HouseSizeSqft']])
prediction
```
