```python
%matplotlib inline
import matplotlib.pyplot
import pandas
import seaborn
import numpy
```

```python
sat_scores = pandas.read_csv('sat_scores.csv',index_col=0)
sat_scores
```

```python
matplotlib.pyplot.boxplot([
                           sat_scores.Verbal,
                           sat_scores.Math
                           ],labels=['Verbal', 'Math'],
                           widths=[0.5,0.5]
                           )
```

```python
low_rate = sat_scores[sat_scores.Rate<10]
low_rate.copy()
```

```python
matplotlib.pyplot.boxplot([
                           low_rate['Verbal'],
                           low_rate['Math']
                           ],labels=['Verbal', 'Math'],
                           widths=[0.5,0.5]
                           )
```

```python
fig, axes = matplotlib.pyplot.subplots(1,2, figsize=(16,10))
sat_scores[['Verbal', 'Math']].plot.barh(ax=axes[0])
low_rate[['Verbal', 'Math']].plot.barh(ax=axes[1])
fig, axes
```
