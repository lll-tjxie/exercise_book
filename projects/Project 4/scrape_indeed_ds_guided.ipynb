```python
# load the library
from bs4 import BeautifulSoup
import urllib2, requests
import pandas as pd

df = pd.DataFrame()   # create a new data frame
pd.set_option('max_colwidth',500)    # to remove column limit (default is 50. Otherwise, we'll lose some info)


# indeed.com's query url
base_url = 'http://www.indeed.com/jobs?q=data+scientist'
#('http://www.indeed.com/jobs?q=data+scientist  ()  &jt=fulltime')
sort_by = 'date'          # sort by data
start_from = '&start='    # start page number
```

```python

#"http://www.indeed.com/jobs?q=data+scientist+$50,000&jt=fulltime"
salary_range = []
Salary_Estimate = [50000,70000,85000,100000,120000]
for salary in Salary_Estimate:
    url = base_url+'+$'+str(salary)+'&jt=fulltime'
    for page in range(1,11):  # Increase number.
        page = (page-1) * 10  
        url_x = url+start_from+str(page) # get full url 
        contents = BeautifulSoup(urllib2.urlopen(url_x), "lxml") 
        contentsElements = contents.find_all('div', attrs={'class' : '  row  result'})
        job_last = contents.find('div',attrs={'class':"lastRow  row  result"})
        contentsElements.append(job_last)
        
        
    
        for elem in contentsElements: 
            try:
                company = elem.find('span', attrs={'itemprop':'name'}).getText().strip() #strips off whitespace 
            except AttributeError:
                company = '-'
            job_title = elem.find('a', attrs={'class':'turnstileLink'}).get('title')
            home_url = "http://www.indeed.com"
            job_link = "%s%s" % (home_url,elem.find('a').get('href'))
            job_addr = elem.find('span', attrs={'itemprop':'addressLocality'}).getText()
            job_posted = elem.find('span', attrs={'class': 'date'}).getText()
        
            try:
                job_review = elem.find('span', attrs={'class':'slNoUnderline'}).getText()
            except AttributeError: 
                job_review = '-'
            
            
            salary = elem.find('td', attrs={'class':'snip'})
            try:
                job_salary = salary.find('nobr').getText.strip()
            except AttributeError:
                job_salary = '-'

            if '50000' in url_x:
                salary_range.append(50000)
                continue
            if '70000' in url_x:
                salary_range.append(70000)
                continue
            if '85000' in url_x:
                salary_range.append(85000)
                continue
            if '100000' in url_x:
                salary_range.append(100000)
                continue
            if "120000" in url_x:
                salary_range.append(120000)
                continue
            else:
                salary_range.append('-')
                continue
    
            df = df.append({'company': company, 
                        'job_title': job_title, 
                        'job_link': job_link, 
                        'job_posted': job_posted,
                        'job_location': job_addr,
                        'job_salary': job_salary,
                        'job_review': job_review,
                       }, ignore_index=True)
     
    
    
    

df

    
```
