```python
# Friday version
total_pronoun_count = []
total_noun_count = []
total_word_count = []
for book_id in range(1661, 1678):
    #print book_id
    with open('pg' + str(book_id) +'.txt') as f:
        one_book = f.read().decode('utf-8',errors='ignore')
        pronoun_count = 0
        noun_count = 0
        word_count = 0
        sentences = nltk.sent_tokenize(one_book)
        words = nltk.word_tokenize(one_book)
        for sentence in sentences:
            tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
            for (word, part_of_speech) in tagged:
                word_count += 1
                if part_of_speech == 'NN':
                    noun_count += 1
                if part_of_speech == 'NNP':
                    pronoun_count += 1
    total_pronoun_count.append(pronoun_count)
    total_noun_count.append(noun_count)
    total_word_count.append(word_count)
```
