```python
populations = dfs[1].copy()
populations
```

```python
#delete useless columns
del populations['Rank']
del populations['House Seats']
del populations['Elect. Votes']
del populations["Pop. per House Seat"]
del populations['Pop. per Elect. Vote']
del populations['Pop. per Senate Seat']
populations
```

No abbreviation! 

```python
wiki_page_abbr = 'https://en.wikipedia.org/wiki/List_of_U.S._state_abbreviations#Postal_codes'
dfs_abbr = pandas.read_html(wiki_page_abbr)
```

```python
dfs_abbr[0].shape
```

```python
dfs_abbr[1].shape
```

```python
dfs_abbr = pandas.read_html(wiki_page_abbr, header=2)
dfs_abbr[0]
```

```python
abbr = dfs_abbr[0].copy()
abbr
```

```python
del abbr['US USA 840']
del abbr['00']
del abbr['Unnamed: 5']
del abbr['Unnamed: 6']
del abbr['U.S.']
del abbr['U.S..1']
del abbr['U.S.A.']
abbr
```

```python
populations_abbr = populations.merge(abbr, left_on='State', right_on='United States of America', how = 'left')
populations_abbr
```

To Merge sat_scores.csv

```python
sat_scores = pandas.read_csv('sat_scores.csv')
sat_scores
```

```python
complete_table_missing_DC = populations_abbr.merge(sat_scores,left_on='US',right_on='State', how = 'right')
complete_table_missing_DC
```

Look for population of DC
Put it into table

```python
complete_table_missing_DC = complete_table_missing_DC[0:50]
complete_table_missing_DC
```

To create a DataFrame with populationm, rate and scores of DC

```python
s = pandas.DataFrame(data = {
        'State_x':['District of Columbia'],
        'Population':[672228.0],
        'United States of America':['District of Columbia'],
        'Federal state': ["State"],
        'US': ['DC'],
        'State_y':["DC"],
        'Rate':[56],
        'Verbal':[482],
        'Math':[474]
    }
)

complete_table1 = complete_table_missing_DC.append(s)
```

```python
complete_table1
```

```python
del complete_table1['Federal state']
del complete_table1['State_x']
del complete_table1['State_y']
complete_table1
```

```python
sorted_by_rate = complete_table1.sort('Rate', ascending = False)
sorted_by_rate
```

To make sorted_by_rate table into 2 parts.
one is for states with high participation rate
one is for states with low rate

```python
high_rate = sorted_by_rate[0:14].sort("Rate",ascending = False)
high_rate
```

```python
low_rate = sorted_by_rate[37:52].sort("Rate",ascending = False)
low_rate
```

```python
high_rate.Population.sum()
```

```python
low_rate.Population.sum()
```

```python
high_rate.Population.sum() - low_rate.Population.sum()
```
